```python
import os
from PyPDF2 import PdfReader, PdfWriter

def extract_pdf_pages():
    # 修正路径
    input_path = r"D:\暮迟\学习\数信\6-数学信号.pdf"
    output_dir = r"D:\暮迟\学习\数信"
    output_path = f"{output_dir}/信号上答案_400-415页.pdf"
    
    # 创建目录
    os.makedirs(output_dir, exist_ok=True)
    
    try:
        with open(input_path, "rb") as infile:
            reader = PdfReader(infile)
            writer = PdfWriter()
            
            # 提取页码
            for page_num in range(400, 415):
                if page_num < len(reader.pages):
                    writer.add_page(reader.pages[page_num])
            
            # 写入新文件（使用wb模式确保完整写入）
            with open(output_path, "wb") as outfile:
                writer.write(outfile)
        
        print(f"成功生成：{output_path}")
        return True
    
    except Exception as e:
        print(f"错误发生：{str(e)}")
        return False

if __name__ == "__main__":
    extract_pdf_pages()
```

```python
import os
from PyPDF2 import PdfReader, PdfWriter

def batch_extract_pdf_pages():
    #路径
    #input_path = r"D:\暮迟\实习\上海东方激光\重庆\正文.pdf"
    #output_dir = r"D:\暮迟\实习\上海东方激光\重庆\截断"
    
    os.makedirs(output_dir, exist_ok=True)
    
    try:
        with open(input_path, "rb") as infile:
            reader = PdfReader(infile)
            total_pages = len(reader.pages)
            
            #提取页码范围
            page_ranges = [(start, start+3) for start in range(13, 16,3)]
            
            for start, end in page_ranges:
                if start >= total_pages:
                    break
                    
                #PDF写入器
                writer = PdfWriter()
                
                for page_num in range(start, min(end, total_pages)):
                    writer.add_page(reader.pages[page_num])
                
                #输出文件名
                output_path = f"{output_dir}/知识_{start}-{end}页.pdf"
                
                #写入文件
                with open(output_path, "wb") as outfile:
                    writer.write(outfile)
                
                print(f"成功生成：{output_path}")
        
        return True
    
    except Exception as e:
        print(f"错误发生：{str(e)}")
        return False

if __name__ == "__main__":
    batch_extract_pdf_pages()
```

```python
import os
import openpyxl
from openpyxl.utils import get_column_letter
from tqdm import tqdm

# 目录设置
#input_dir = r"D:\暮迟\实习\上海东方激光\湖南\临时知识（1.1-7.2）"
#output_dir = r"D:\暮迟\实习\上海东方激光\湖南\细目表\知识梳理(1.1~7.2)"

# 确保输出目录存在
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
    print(f"已创建输出目录: {output_dir}")

# 定义公式字典
formulas = {
    'E': '=IF(H{row}="",I{row},IF(I{row}="",H{row},IF(H{row}=I{row},H{row},H{row}&"、"&I{row})))',
    'F': '=IF(L{row}="","",LEFT(L{row},FIND("@",SUBSTITUTE(L{row},"-","@",4))-1))',
    'G': '=IF(M{row}="","",LEFT(M{row},FIND("@",SUBSTITUTE(M{row},"-","@",4))-1))',
    'H': '=IF(F{row}="","",LEFT(F{row},FIND("-",F{row},FIND("-",F{row})+1)-1))',
    'I': '=IF(G{row}="","",LEFT(G{row},FIND("-",G{row},FIND("-",G{row})+1)-1))',
    'J': '=IF(F{row}=G{row},"",G{row})',
    'X': '=IF(COUNTBLANK(L{row}:N{row})>=2,1,2)'
}

# 批量处理文件
for file_name in tqdm(os.listdir(input_dir), desc="处理文件中"):
    if file_name.endswith(('.xlsx', '.xls')):
        input_path = os.path.join(input_dir, file_name)
        output_path = os.path.join(output_dir, f"f_{file_name}")
        
        try:
            # 打开Excel并保留原格式
            wb = openpyxl.load_workbook(input_path)
            ws = wb.active
            
            # 从第3行开始填充公式
            for row in range(3, ws.max_row + 1):
                for col_letter, formula_template in formulas.items():
                    cell = f"{col_letter}{row}"
                    # 只填充空单元格或覆盖已有内容（根据需求调整）
                    if ws[cell].value is None or isinstance(ws[cell].value, str):
                        ws[cell] = formula_template.format(row=row)
            
            # 保存并保留原格式
            wb.save(output_path)
            
            # 显示第一个文件处理结果示例
            if file_name == os.listdir(input_dir)[0]:
                print(f"\n示例文件 {file_name} 处理结果：")
                sample_row = 3  # 查看第3行
                for col in sorted(formulas.keys()):
                    print(f"{col}{sample_row}: {ws[f'{col}{sample_row}'].value}")
                
        except Exception as e:
            print(f"\n处理失败: {file_name} - {str(e)}")

print("\n所有文件公式添加完成！")
print(f"处理后的文件已保存至: {output_dir}")
```

```python
import pandas as pd
import os

def lookup_knowledge_content_interactive():
    # 文件路径
    excel_path = r"D:\暮迟\实习\上海东方激光\重庆\20250826重庆高职考字典库(1).xlsx"
    #excel_path = r"D:\暮迟\实习\上海东方激光\广东\20250815广东应试数学字典库.xlsx"
    
    try:
        # 检查文件是否存在
        if not os.path.exists(excel_path):
            raise FileNotFoundError(f"文件未找到: {excel_path}")
        
        # 读取所有工作表
        xl = pd.ExcelFile(excel_path)
        all_dfs = {}
        for sheet_name in xl.sheet_names:
            df = xl.parse(sheet_name)
            # 自动检测列名（兼容H/I列或第8/9列）
            if len(df.columns) >= 9:
                code_col = df.columns[7]  # 第8列
                content_col = df.columns[8]  # 第9列
                # 创建代码到内容的映射
                code_map = {}
                for _, row in df.iterrows():
                    code_value = str(row[code_col]).strip()
                    if code_value:  # 忽略空代码
                        code_map[code_value] = row[content_col]
                all_dfs[sheet_name] = code_map
        
        print("="*50)
        print("知识内容查询系统（输入空行结束）")
        print(f"已加载 {len(all_dfs)} 个工作表")
        print("="*50)
        
        # 创建所有代码的全局映射
        global_code_map = {}
        for sheet_map in all_dfs.values():
            global_code_map.update(sheet_map)
        
        while True:
            # 获取用户输入
            print("\n请粘贴或输入需要查询的代码（支持单行多个代码，用空格分隔）：")
            input_line = input().strip()
            if not input_line:
                print("检测到空输入，程序结束。")
                break
            
            # 分割输入（兼容空格分隔的多个代码）
            codes = []
            for item in input_line.split():
                # 清理代码：移除可能的前后空格
                clean_code = item.strip()
                # 只添加非空代码
                if clean_code:
                    codes.append(clean_code)
            
            # 查询并输出结果
            print("\n查询结果：")
            found_all = True
            results = []
            
            for code in codes:
                # 尝试在全局映射中查找代码
                if code in global_code_map:
                    results.append(f"{code.ljust(20)}{global_code_map[code]}")
                else:
                    results.append(f"{code.ljust(20)}[未找到匹配内容]")
                    found_all = False
            
            # 打印结果
            print("\n".join(results))
            if not found_all:
                print("\n警告：部分代码未匹配到内容，请检查输入！")
                print(f"提示：当前字典库中共有 {len(global_code_map)} 个唯一代码")
    
    except FileNotFoundError as e:
        print(f"错误：{str(e)}")
    except Exception as e:
        print(f"程序出错: {str(e)}")
        import traceback
        traceback.print_exc()

# 运行交互式查询
if __name__ == "__main__":
    lookup_knowledge_content_interactive()
```
